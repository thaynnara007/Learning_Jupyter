```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils
from keras.datasets import mnist
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
base = datasets.load_iris()
previsores = base.data
previsores
```

```python
classe = base.target
classe
```

```python
# temos 3 classes (0,1,2), logo serão 3 neurônios de saída na rede neural
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
x_treinamento, x_test, y_treinamento, y_test = train_test_split(previsores, classe_dummy, test_size=0.3, random_state=0)
modelo = Sequential()
modelo.add(Dense(units=5,input_dim=4))
modelo.add(Dense(units=4))
modelo.add(Dense(units=3, activation = 'softmax'))
```

```python
y_test
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) 
```

```python
modelo.fit(x_treinamento, y_treinamento, epochs=1000, validation_data=(x_test,y_test))
```

```python
previsoes = modelo.predict(x_test)
previsoes
```

```python
previsoes = (previsoes > 0.5)
previsoes
```

```python
# para fazer a matriz de confusão
y_test_matriz = [np.argmax(t) for t in y_test]
y_test_matriz
```

```python
y_previsao_matriz = [np.argmax(t) for t in previsoes]
y_previsao_matriz
```

```python
confusao = confusion_matrix(y_test_matriz, y_previsao_matriz)
confusao
```

```python
(x_treinamento,y_treinamento), (x_test, y_test) = mnist.load_data()
print (x_treinamento.shape)
x_treinamento.dtype
```

```python
plt.imshow(x_treinamento[1], cmap='gray')
plt.title(y_treinamento[1])
```

```python
plt.imshow(x_treinamento[20], cmap='gray')
plt.title(y_treinamento[20])
```

```python
x_treinamento = x_treinamento.reshape((len(x_treinamento), np.prod(x_treinamento.shape[1:])))
x_treinamento.shape
```

```python
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
x_test.shape
```

```python
x_treinamento = x_treinamento.astype('float32')
x_treinamento.dtype
```

```python
x_test = x_test.astype('float32')
x_test.dtype
```

```python
x_treinamento /= 255
x_treinamento[1]
```

```python
x_test /= 255
x_test[1]
```

```python
y_treinamento = np_utils.to_categorical(y_treinamento,10)
y_test = np_utils.to_categorical(y_test,10)
```

```python
modelo = Sequential()
modelo.add(Dense(units=64, activation='relu', input_dim=784))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=64, activation='relu'))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=64, activation='relu'))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=10, activation='softmax'))
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
historico = modelo.fit(x_treinamento, y_treinamento, epochs=20, validation_data=(x_test, y_test))
```

```python
historico.history.keys()
```

```python
plt.plot(historico.history['val_loss'])
```

```python
plt.plot(historico.history['val_acc'])
```
